# Contextualizing my GPA- Zara

### I used the codes below to create a series of graphics to show my grades during a given semester and the credits I took during that semester. Plot 1 was my first draft and although I'm continuing to update and create better drafts, at the moment, I believe Plot 3 is the most visually pleasing representation and that is the one which I most frequently use. All of these plots are created using open-sourced python libraries so they are quite simple to make. Feel free to use the codes or the "idea" of this graphic below :-)

```python
##importing the necessary libraries 

import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
from  matplotlib import colors
import matplotlib.cm as cmap
import matplotlib.colors as mcolors
```

```python
## creating arrays of semesters (x), corresponding term grade (y), and final GPA

x = ["Fl '19", "Sp '20", "Fl '20", "Sp '21", "Fl '21", "Sp '22", "Su '22", "Fl '22", "Sp '23"]
y = [3.650, 4.000, 2.445, 3.350, 3.238, 2.847, 4.000, 3.441, 3.552]
z = []
for items in x:
    z.append(3.300)
c = [12, 7, 15.5, 12, 13, 17, 6, 17, 23]
d = [3.30, 3.30, 3.30, 3.30, 3.30, 3.30, 3.30, 3.30, 3.30]
```

```python
## draft 1

fig, axis = plt.subplots()

axis.plot(x,y, label = "Semester GPA", marker = 'o', color= 'steelblue')
axis.plot(x, z, label= "Final Cumulative GPA", color = 'darkgreen')
axis.set_xlabel("College semesters from Fall 2019 to Spring 2023. Key: Fl = Fall, Sp = Spring, & Su = Summer", fontfamily= "AppleMyungjo")

axis.fill_between(x, y, z, alpha = 0.2, color = 'red')
axis.axvline(x=2, color='red', linestyle='dashed', alpha= 0.5)
axis.axvline(x=5, color='red', linestyle='dashed', alpha= 0.5)

axis.annotate("Semesters where\nI was facing housing\ninsecurity and\ncontinued to work\nmy full-time job", xy= (2.20,2.25), xytext= (2.65,2), fontsize= 8, fontfamily= "AppleMyungjo")
for i, txt in enumerate(y):
    axis.annotate(txt, (x[i],y[i]), xytext= (x[i],y[i]+0.07), color= 'black', fontsize = 5)

axis.set_ylim([0,4.2])
axis.set_ylabel('GPA on a 4.0 scale', fontfamily= "AppleMyungjo")

axis2 = axis.twinx()
axis2.set_ylabel('College Credits', fontfamily= "AppleMyungjo")
axis2.bar(x,c, alpha= 0.4, color= 'green', width= 0.2, label = "Credits Taken")
for i, txt in enumerate(c):
    axis2.annotate(txt, (x[i],c[i]), xytext= (x[i],y[i]), color= 'black', fontsize = 5, textcoords="offset points", va= 'top', ha = 'right')
axis2.set_ylim([0,30])

fig.legend(loc= 'upper right', fontsize = 7)
plt.title("Contextualizing GPA- Zara", fontfamily= "AppleMyungjo")
plt.show

#commenting the lines out below because I do not wish to save this graphs on my computer 

#save_to = '/Users/zaraafrooz/Desktop/'
#plt.savefig(save_to + 'Contextualizing_my_GPA_1.png', dpi= 300)
```

```python
## draft 2
## instead of creating the graphic with double sided y-axis, create a subplot with one shared y-axis for a cleaner presentation.

fig, axis = plt.subplots(ncols = 1, nrows=2, sharex= True)
plt.subplots_adjust(hspace = 0.0, wspace = 0.0)

axis[0].plot(x,y, label = "Semester GPA", marker = 'o', color= 'steelblue')
axis[0].plot(x,z, label= "Final Cumulative GPA: 3.30", color = 'darkgreen') #linestyle = 'dashed') 
axis[1].bar(x,c, alpha= 0.4, color= 'green', width= 0.2, label = "Credits Taken")
axis[1].set_ylabel('College Credits', fontfamily= "AppleMyungjo")


axis[1].set_xlabel("Semesters from Fall 2019 to Spring 2023: Fl = Fall, Sp = Spring, & Su = Summer", fontfamily= "AppleMyungjo")

axis[0].set_ylim(2.2,4.3)
axis[0].set_yticks([2.5, 3, 3.5, 4])
for i, txt in enumerate(y):
    if y[i]>3.30:
        axis[0].annotate(txt, (x[i],y[i]), xytext= (x[i],y[i]+0.1), color= 'black', fontsize = 10, ha="center", va="center")
    else:
        axis[0].annotate(txt, (x[i],y[i]), xytext= (x[i],y[i]-0.15), color= 'black', fontsize = 10, ha="center", va="center")


axis[1].set_ylim(0,26.5)
axis[1].set_yticks([0, 6, 12, 18, 24])
for i, txt in enumerate(c):
    axis[1].annotate(txt, (x[i],c[i]), xytext= (x[i],c[i]+2), color= 'black', fontsize = 10, ha="center", va="center")

axis[0].margins(x=0.05)
axis[1].margins= axis[0].margins

axis[0].fill_between(x, y, z, alpha = 0.2, color = 'red')
axis[0].set_ylabel('GPA on a 4.0 scale', fontfamily= "AppleMyungjo")

axis[0].axvline(x=2, color='red', linestyle='dashed', alpha= 0.5)
axis[0].axvline(x=5, color='red', linestyle='dashed', alpha= 0.5)
axis[1].axvline(x=2, color='red', linestyle='dashed', alpha= 0.5)
axis[1].axvline(x=5, color='red', linestyle='dashed', alpha= 0.5)

axis[1].annotate("Semesters where\nI was facing housing\ninsecurity and\ncontinued to work\nmy full-time job", xy= (2.20,2.25), xytext= (2.65,18), fontsize= 8, fontfamily= "AppleMyungjo")

fig.legend(loc= 'upper right', fontsize = 7)
fig.suptitle("Contextualizing GPA- Zara", fontfamily= "AppleMyungjo")


## save_to = '/Users/zaraafrooz/Desktop/'
## plt.savefig(save_to + 'Contextualizing_my_GPA_2.png', dpi= 300)
```
